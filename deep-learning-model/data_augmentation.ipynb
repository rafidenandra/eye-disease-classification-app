### Data augmentations using OpenCV

```python
import cv2
import os

def h_flip(image, image_file):
    # Apply horizontal flip
    h_flipped_image = cv2.flip(image, 1)
    h_flipped_filename = 'hflip_' + image_file
    h_flipped_image_path = os.path.join(output_dir, h_flipped_filename)
    cv2.imwrite(h_flipped_image_path, h_flipped_image)

def v_flip(image, image_file):
    # Apply vertical flip
    v_flipped_image = cv2.flip(image, 0)
    v_flipped_filename = 'vflip_' + image_file
    v_flipped_image_path = os.path.join(output_dir, v_flipped_filename)
    cv2.imwrite(v_flipped_image_path, v_flipped_image)

def rotation_range(image, image_file, min, max):
    # Apply rotation range of -20 to 20 degrees
    for angle in range(min, max):
        rotated_image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
        rotated_filename = f'rotate_{angle}_' + image_file
        rotated_image_path = os.path.join(output_dir, rotated_filename)
        cv2.imwrite(rotated_image_path, rotated_image)


def augment_images(input_dir, output_dir):
    # Create the output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)
    
    # List the class directories in the input directory
    image_files = os.listdir(input_dir)
    
    # Iterate over the image files
    for image_file in image_files:
        # Load the image
        image_path = os.path.join(input_dir, image_file)
        print('image path:', image_path)
        image = cv2.imread(image_path)
        
        # Resize the image to 100x100 pixels
        resized_image = cv2.resize(image, (100, 100))
        
        # h_flip(resized_image, image_file)
        v_flip(resized_image, image_file)
        # rotation_range(resized_image, image_file, -20, 20)

# Example usage
input_dir = 'D:\\Kuliah\\Proyek Akhir 2022\\Dataset\\Dataset\\Normal'
output_dir = 'mixed_augmentations_examples\\Normal'
augment_images(input_dir, output_dir)
```

### Data augmentations using ImageDataGenerator

```python
data_dir = 'D:\\Kuliah\\Proyek Akhir 2022\\Dataset\\Dataset\\Normal'
augmented_dir = 'mixed_augmentations_examples\\Normal'

# Define the preprocessing techniques and their respective prefixes
augmentation_prefixes = {
    'rotation_range': 'rotated_'
}

# Create the data generator with the desired preprocessing techniques
data_generator = ImageDataGenerator(
    rotation_range=20,
    fill_mode='nearest'
)

def load_and_preprocess_image(image_path):
    img = load_img(image_path, target_size=(100, 100))  # Load image and resize it to the desired shape
    img = img_to_array(img)  # Convert image to array and normalize pixel values to the range of 0-1
    return img

image_filenames = os.listdir(data_dir)

# Iterate over each image filename
for filename in image_filenames:
    image_path = os.path.join(data_dir, filename)

    # Load and preprocess the image
    img = load_and_preprocess_image(image_path)

    # Generate augmented images using the data generator
    augmented_images = data_generator.flow(
        img[np.newaxis, ...],
        batch_size=1,
        save_to_dir=augmented_dir,
        save_prefix='',  # Save prefix will be determined by the augmentation technique
        save_format='jpg'
    )

    # Generate and save the augmented images
    for augmentation_name, augmentation_prefix in augmentation_prefixes.items():
        augmented_images.save_prefix = augmentation_prefix

        # Generate and save the specified number of augmented images
        for i in range(1):
            augmented_image = augmented_images.next()[0].astype('uint8')
            augmented_image_filename = augmentation_prefix + filename
            augmented_image_path = os.path.join(augmented_dir, augmented_image_filename)
            Image.fromarray(augmented_image).save(augmented_image_path)
```
