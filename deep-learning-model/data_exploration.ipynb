```python
# Show dataset detail

def show_dataset_detail(dataset):
    dataset_pos = dataset[dataset['types'] == 1]
    dataset_neg = dataset[dataset['types'] == 0]
    print('Positive data:', len(dataset_pos))
    print('Negative data:', len(dataset_neg))

# Copy dataset into dataset temp folder

dest_glaucoma_path = 'G:/Proyek Akhir/Dataset Temp/Glaucoma - 3. Mixed Glaucoma Database Dataset/Glaucoma'
dest_normal_path = 'G:/Proyek Akhir/Dataset Temp/Glaucoma - 3. Mixed Glaucoma Database Dataset/Normal'
source_path = 'G:/Proyek Akhir/Dataset Holder/\Glaucoma - 3. Mixed Glaucoma Database Dataset/full-fundus/'

def copy_dataset(dataset):
    dataset_pos = dataset[dataset['types'] == 1]
    dataset_neg = dataset[dataset['types'] == 0]

    for file_path_pos in dataset_pos['fundus']:
        source = source_path + file_path_pos
        destination = dest_glaucoma_path
        shutil.copy2(source, destination)
        print('copied', file_path_pos, 'successfully')

    for file_path_neg in dataset_neg['fundus']:
        source = source_path + file_path_neg
        destination = dest_normal_path
        shutil.copy2(source, destination)
        print('copied', file_path_neg, 'successfully')
```

```python
show_dataset_detail(HRF)
copy_dataset(HRF)
```
